### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset('tips') # inbuilt data
tips
```

```python
# Scatterplot --> axes level function
# sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue='sex', style='time', size='size') #in seaborn (hue) is used to know how histogram differs between two categories

# relplot --> figure level function (usually, square shape)  (in general we have to use the figure level function as it provides general function, where we just change the kind type and different types of graphs can be plot)
# sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
sns.relplot(data=tips, x='total_bill', y='tip',kind='scatter', hue='sex', style='smoker', size='size')
```

## LinePlot
- it is same as the scatter plot, but has line connected for the continuation of the data, like in the x-axis we can have time etc

```python
# Line Plot

gap = px.data.gapminder() # inbuilt data
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India', 'Germany', 'Brazil'])]
temp_df
```

```python
# figure level relplot

sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# axis level

sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent')
```

### 2. Distribution Plots (Used for the distribution of the data i.e Divided)

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency (mean, median, centere)
- is the data bimodal? (multiple peaks)
- Are there outliers?

Plots under distribution plot

- histplot (Data Distribution by creating the bins)
- kdeplot ()
- rugplot

```python
# figure level --> displot
# axes level --> histplot, kdeplot, rugplot
```

## 1. histplot()

```python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='hist', bins=30)
```

# Bins parameter customization

```python
sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countPlot (shows the count of the values of the categories of the categorial values)
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameters
# sns.displot(data=tips, x='tip', kind='hist', hue='sex')

# element --> step (for better reading convention)
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element= 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', hue='sex')
```

```python
# faceting using col and row --> will not work on histplot fucntion
sns.displot(data=tips, x='tip', kind='hist', col='sex', element='step')
```

## 2. kdeplot()

```python
# kernel density estimate (kde) function
# rather than using discrete bins, a kde plot smooths the observations with a Gaussian Kernel, producing a continous density estimate, in simple words it smoothens the curve rather than discrete bins

# axes level function
sns.kdeplot(data=tips, x='total_bill')

```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue --> fill
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

## rugplot()

```python
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrucsive way.
# It shows the same distribution like histogram but in differnt way like, horizontally

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```
