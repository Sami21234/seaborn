## 2. kdeplot()

```python
# kernel density estimate (kde) function
# rather than using discrete bins, a kde plot smooths the observations with a Gaussian Kernel, producing a continous density estimate, in simple words it smoothens the curve rather than discrete bins

# axes level function
sns.kdeplot(data=tips, x='total_bill')

```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue --> fill
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

## rugplot()

```python
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrucsive way.
# It shows the same distribution like histogram but in differnt way like, horizontally

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

# The main difference between the strip and swarm plot is that ->
- strip - simple scatter, may overlap the data points, It is widely used for medium to large dataset
- swarm - smarter scatter, no overlap, It is widely used for small dataset

It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

### Categorial Estimate Plot -> for central tendency (for estimating the values like --> mean, median, mode, standard deviation etc, like aggregation)

```python
# 1. Barplot --> Barplot means finding the aggregate value from categorial column to a numerical column

sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='Set1')
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='Set1', errorbar=None)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# Now applying different estimation value to the barplot using (estimator)

import numpy as np

sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='Set1', estimator=np.min)
```

### Pointplot (same as barplot but without bars, it basically connects the bars)

```python
# pointplot

sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='Set1', estimator=np.max)
# sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='Set1', estimator=np.max, errorbar=None)
```

### countplot

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# countplot --> It shows the count(frequency) of each category in a categorial variable.

sns.countplot(data=tips, x='sex', hue='smoker')
```

```python
# Now faceting using catplot
sns.catplot(data=tips, x='sex', y='total_bill', col='smoker', kind='box', row='time', hue='sex')
```

### Regression Plots

- regplot (axes level function)
- lmplot (figure level function)

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# Regression plot --> these plots are used to plot the linear regression

# 1. regplot (axes level function) --> regplots draws the scatterplot and also provides the linear regression line
# hue parameter is not possible in axes level
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# figure level function for regression plot
# lmplot

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

### residual plot - A residual plot shows the error in a regression model.
Residual = Actual value - Predicted value,
It tells how far the model's prediction is from the real data.


```python
# residplot
sns.residplot(data=tips, x='total_bill', y='tip')
```

## MultiGrid Plots

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot, displot, catplot, lmplot

# facetplot using row and col.
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin', col='day', row='time', hue='sex', palette='Set1')
```

```python
# and Now using Facetgrid --> It is same as facetplot but give more reliable customizations

g = sns.FacetGrid(data=tips, hue='smoker', col='day', row='time', palette='Set1', margin_titles=True, height=5)
g.map(sns.boxplot, 'sex', 'total_bill')
g.add_legend()
```

## Plotting Pairwise Relationship (PairGrid Vs Pairplot)

### so the main difference is that, the -->
- pairplot provides limited customizations to the different plots,
- pairGrid provides more flexibility and multiple plot customizations

```python
# pairplot creates a pairwise relationship
sns.pairplot(data=iris, hue='species', palette='Set1')
```

```python
# Pairgrid plot (creates a grid layout)
# pairgrid plot creates a detailed pairwise relationship grids and cusomizations like upper, lower triangular and diagonal

g = sns.PairGrid(data=iris, hue='species', palette='Set1')

# upper triangular
g.map_upper(sns.scatterplot)

# lower triangular
g.map_lower(sns.histplot)

# Diagonal
g.map_diag(sns.kdeplot, fill=True)

# Adding legend
g.add_legend()
```

```python
# vars
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

## JointGrid Vs JointPlot

```python
# joinplot --> It is used to study both 1D and 2D at the same time
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde', hue='smoker', fill=True)
```

```python
# jointGrid --> It is more reliable and has more customisation plots

g = sns.JointGrid(data=tips, x='total_bill', y='tip', hue='smoker')
g.plot(sns.scatterplot, sns.violinplot)
```
