### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset('tips') # inbuilt data
tips
```

```python
# Scatterplot --> axes level function
# sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue='sex', style='time', size='size') #in seaborn

# relplot --> figure level function (usually, square shape)  (in general we have to use the figure level function as it provides general function, where we just change the kind type and different types of graphs can be plot)
# sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
sns.relplot(data=tips, x='total_bill', y='tip',kind='scatter', hue='sex', style='smoker', size='size')
```

## LinePlot
- it is same as the scatter plot, but has line connected for the continuation of the data, like in the x-axis we can have time etc

```python
# Line Plot

gap = px.data.gapminder() # inbuilt data
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India', 'Germany', 'Brazil'])]
temp_df
```

```python
# figure level relplot

sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# axis level

sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent')
```
